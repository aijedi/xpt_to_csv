```python
# Function to convert an .XPT file to .csv format
def xpt_to_csv(filepath, savedir):
    with open(filepath, 'rb') as f:
        df = xport.to_dataframe(f)
    print (df.columns.values)
    name = filepath.split('.')[0]
    savepath = save_dir +'/' + name + '.csv'
    df.to_csv(savepath)
    
    return True
```
